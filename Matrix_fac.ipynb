```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

# Manipulating the data to use

```python
df=pd.read_csv("out.csv", encoding ="utf8", engine="python")
df=df.drop(columns=['pf_o_att','pf_o_sin','pf_o_int','pf_o_fun','pf_o_amb','pf_o_sha','race','age'])
df_pivot = pd.pivot_table(df, values='like_o', index=['iid','go_out','sports','tvsports','exercise',
'dining','museums','art','hiking','gaming','clubbing','reading','tv','theater','movies','concerts','music',
'shopping','yoga'], columns= 'pid',aggfunc=np.sum, fill_value=0)
df_pivot.head()
```

# Females

```python
females = df.groupby(['gender']).get_group(0)
females= females.drop(females.columns[0], axis = 1)
females = females.fillna(0)
rating_count_f = females.groupby(by=['iid'])['pid'].count().reset_index().rename(columns={'pid':'totalPartnerCount'})
[['iid','pid']]
females = females.merge(rating_count_f, left_on='iid',right_on='iid')
```

```python
#Females stats based on their popularity (more partner's ratings)
females['totalPartnerCount'].describe()
```

```python
#Quantiles
females['totalPartnerCount'].quantile(np.arange(.9,1,.01))
```

```python
females = pd.pivot_table(females, values='like_o', index=['iid','go_out','sports','tvsports','exercise',
'dining','museums','art','hiking','gaming','clubbing','reading','tv','theater','movies','concerts','music',
'shopping','yoga'], columns= 'pid',aggfunc=np.sum, fill_value=0)
females.head()
```

```python
X=females.values.T
X.shape
```

```python
#SVD
from sklearn.decomposition import TruncatedSVD

SVD = TruncatedSVD(n_components=12,random_state=17)
matrix_X = SVD.fit_transform(X)
matrix_X.shape
```

```python
import warnings
warnings.filterwarnings("ignore",category=RuntimeWarning)
corr = np.corrcoef(matrix_X)
np.isfinite(corr)
corr.shape
```

```python
corr_1=corr[1]
corr_1
```

# Best Match

```python
partners = females.columns
match=list(partners[(corr_1<1)&(corr_1>.90)])
match
```

# Average Match

```python
partners = females.columns
match_avg=list(partners[(corr_1<.89)&(corr_1>.75)])
match_avg
```

# Poor Match

```python
partners = females.columns
no_match=list(partners[(corr_1<.25)&(corr_1>.01)])
no_match
```
